```python

import pandas as pd
import numpy
import tensorflow as tf


from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

```

```python
# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.set_random_seed(seed)
```

```python
base_url="https://raw.githubusercontent.com/greencodespace/data_deeplearning/master/sonar.csv"

df = pd.read_csv(base_url,header=None); #, names = [])
```

```python
#df=pd.read_csv("c:/tmp/Sonar.csv")
```

```python
print(df.head());
```

```python
dataset=df.values;
(rows,cols)=dataset.shape;

#x=dataset[:,0:cols-1];
#y=dataset[:,cols-1];
x=dataset[:,0:60];
y=dataset[:,60];
```

```python
e=LabelEncoder();
e.fit(y);
y1=e.transform(y);      
```

```python
model=Sequential();
model.add(Dense(24, input_dim=60, activation='relu'));
model.add(Dense(10, activation='relu'));
model.add(Dense(1,activation='sigmoid'));
```

```python
model.compile(loss='mean_squared_error',optimizer='adam' ,metrics=['accuracy'])
```

```python
model.fit(x,y1,epochs=200,batch_size=5);
```

```python
print("\n Accuracy %.4f" % (model.evaluate(x,y1)[1]) )
```
